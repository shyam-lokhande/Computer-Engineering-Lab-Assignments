```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Housing.csv')
pd.set_option('display.max_rows',506)
df
```

```python
df.isnull().sum()
```

```python
corr_matrix=df.corr()
corr_matrix['MEDV']
```

```python
model=LinearRegression()
x=df[['INDUS','NOX','RM','TAX','PTRATIO','LSTAT']]
y=df['MEDV']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=85)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,confusion_matrix,mean_squared_error
r2_sc=r2_score(y_test,y_predict)
```

```python
r2_sc
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_predict))
rmse
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.regplot(x=y_test,y=y_predict,data=df)
```
