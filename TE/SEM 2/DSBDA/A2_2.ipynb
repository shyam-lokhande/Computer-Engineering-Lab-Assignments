```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('EduData.csv')
pd.set_option('display.max_rows',480)
df.size
```

```python
df.replace('?',np.nan,inplace=True)
df.isnull().sum()
df
```

```python
df['Topic'].value_counts()
df['Topic'].replace(np.nan,'IT',inplace=True)
```

```python
df.dropna(subset=['GradeID','Relation'],axis=0,inplace=True)
```

```python
rh_avg=df['raisedhands'].astype('float').mean()
df['raisedhands'].replace(np.nan,rh_avg,inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df['NationalITy'].value_counts())
df.drop(df[(df['NationalITy']=='venzuela') | (df['NationalITy']=='Morocco')].index,axis=0,inplace=True)
df['NationalITy'].value_counts()
```

```python
print(df['Semester'].value_counts())
df.drop(df[(df['Semester']=='T')].index,axis=0,inplace=True)
df['Semester'].value_counts()
```

```python
df.drop('ParentAnsweringSurvey',axis=1,inplace=True)
df.head()
```

```python
df['raisedhands']=df['raisedhands'].astype('float')
df['raisedhands']=df['raisedhands']/df['raisedhands'].max()
df['VisITedResources']=df['VisITedResources'].astype('float')
df['VisITedResources']=df['VisITedResources']/df['VisITedResources'].max()
df.head()
```

```python
df['GradeID'].value_counts()
cols={'G-02':1,'G-04':2,'G-05':3,'G-06':4,'G-07':5,'G-08':6,'G-09':7,'G-10':8,'G-11':9,'G-12':10}
df['GradeID']=[cols[i] for i in df.GradeID]
df.head()
```

```python
df['Grades']=pd.cut(df['GradeID'],bins=4,labels=['Worst','Normal','Good','Best'])
```
