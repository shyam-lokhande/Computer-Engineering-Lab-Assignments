```python
import numpy as np
import pandas as pd
import seaborn as sns
import nltk
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
df=pd.read_csv('spam.csv', encoding="ISO-8859-1")
df=df.iloc[:,:2]
df
```

```python
df.isnull().sum()
```

```python
from string import punctuation
def rm_punctuation(text):
    translator=str.maketrans("","",punctuation)
    return text.translate(translator)
df['v2']=df['v2'].apply(rm_punctuation)
```

```python
df['v2']=df['v2'].str.lower()
```

```python
stopwords=nltk.corpus.stopwords.words('english')
def rm_stp(text):
    text=text.split()
    return " ".join(i for i in text if i not in stopwords)
df['rm_stp']=df['v2'].apply(rm_stp)
df
```

```python
df['tags']=df['rm_stp'].apply(lambda x: x.split())
df
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

def lemmatize_text(text):
    lemmatized_text=[]
    for word in text:
        lemmatized_text.append(lemmatizer.lemmatize(word))
    return lemmatized_text

df['lemmatized-text']=df['tags'].apply(lemmatize_text)
df
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

def stem_text(text):
    stemmed_text=[]
    for i in text:
        stemmed_text.append(stemmer.stem(i))
    return stemmed_text

df['stmmed-text']=df['tags'].apply(stem_text)
df
```

```python
# pos tagging
df['pos-tag']=df['lemmatized-text'].apply(lambda x: nltk.pos_tag(x))
df
```

```python
N=len(df)
def df_idf(doc,token):
    word_count=len(df['stmmed-text'][doc])
    counter=dict(nltk.FreqDist(df['stmmed-text'][doc]))
    tf=counter[token]/(word_count+1)
    
    doc_f=sum(1 for i in df['stmmed-text'] if token in i)
    idf=np.log(N/(doc_f+1))
    
    return tf*idf

for i in df['stmmed-text'][0]:
    print(i)
    print(df_idf(0,i))
```
