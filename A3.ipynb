```python
df=pd.read_csv('anti_dataset.csv')
pd.set_option('display.max_rows',40)
df
```

```python
df['EWB']=pd.cut(df['V1'],bins=5,labels=['cat-1','cat-2','cat-3','cat-4','cat-5'])
```

```python
df['EFB']=pd.qcut(df['V1'],q=5)
df
```

```python
df['devided_v1']=pd.cut(df['V1'],bins=2)
df
```

```python
p1=df.iloc[:23,:]
p2=df.iloc[23:,:]
p1
```

```python
plt.boxplot(p1['V1'])
plt.title('Box Plot 1')
plt.ylabel('V1')
plt.show()
```

```python
min_anti_a=p1['ANTI_A'].astype('float').min()
max_anti_a=p1['ANTI_A'].astype('float').max()
p1=p1.copy()
min_anti_b=p1['ANTI_B'].astype('float').min()
max_anti_b=p1['ANTI_B'].astype('float').max()
p1['ANTI_A']=(p1['ANTI_A']-min_anti_a)/max_anti_a-min_anti_a
p1['ANTI_B']=(p1['ANTI_B']-min_anti_b)/max_anti_b-min_anti_b
p1
```

```python
p1.plot.scatter(x='ANTI_A',y='ANTI_B')
```
