```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.isnull().sum()
```

```python
df['age'].replace(np.nan,df['age'].mean(),inplace=True)
df.isnull().sum()
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',style='class' ,data=df)
plt.title('Scatterplot - survival vs sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
m_survived=df[(df['sex']=='male') & df['survived']==1]['age']
f_survived=df[(df['sex']=='female') & df['survived']==1]['age']

plt.hist(np.random.uniform(low=0.0,high=1.0,size=len(m_survived)),label='Male Survival',alpha=0.5)
plt.hist(np.random.uniform(low=0.0,high=1.0,size=len(f_survived)),label='Female Survival',alpha=0.5)
plt.xlabel('uniform Distribution')
plt.ylabel('Count')
plt.title('Histogram')
plt.legend(title='survived',loc='upper right')
plt.show()
```

```python
m_pos=df[(df['sex']=='male') & df['survived']==1]['age']
m_neg=df[(df['sex']=='male') & df['survived']==0]['age']
f_pos=df[(df['sex']=='female') & df['survived']==1]['age']
f_neg=df[(df['sex']=='female') & df['survived']==0]['age']

fig,ax=plt.subplots(1,1,figsize=(10,6))

ax.scatter(np.random.uniform(-0.3, 0.3, len(m_pos)), m_pos, color='blue', edgecolor='black', label='Male(Survived=1)')
ax.scatter(np.random.uniform(-0.3, 0.3, len(m_neg)), m_neg, color='blue', edgecolor='black', alpha=0.2, label='Male(Survived=0)')


# Female Stripplot
ax.scatter(2+np.random.uniform(-0.3, 0.3, len(f_pos)),f_pos, color='green', edgecolor='black', label='Female(Survived=1)')
ax.scatter(2+np.random.uniform(-0.3, 0.3, len(f_neg)), f_neg, color='green', edgecolor='black', alpha=0.2, label='Female(Survived=0)')

plt.title("Age Distribution by sex and Survival")
plt.xlabel("sex")
plt.ylabel("Age")
plt.xlim(-0.5, 2.5)  # set x-axis limits
plt.xticks([0, 1,2], ["Male","", "Female"])  # set x-axis tick labels
plt.legend(title="suvived", loc="upper center")


# show the plots
plt.show()
```
